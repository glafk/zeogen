```python
import pickle

# Load test dataset
with open("combined_test.pickle", "rb") as f:
    data = pickle.load(f)

print(data[0])
```

```python
# Plot HOA per zeolite code
import matplotlib.pyplot as plt

# Plot average HOA per zeolite code
for code in list(set([x["zeolite_code"] for x in data])):
    fig, ax = plt.subplots()
    fig.suptitle(code)
    print([x["hoa"] for x in data if x["zeolite_code"] == code])
    ax.boxplot([x["hoa"] for x in data if x["zeolite_code"] == code])
    plt.show()
```
